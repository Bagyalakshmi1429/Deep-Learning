```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, learning_rate, iterations):
    """
    Perform gradient descent to find the minimum of the given function.

    Parameters:
    - x: Initial guess for the minimum
    - learning_rate: Step size for each iteration
    - iterations: Number of iterations to perform

    Returns:
    - x_min: The value of x at the minimum
    - iterations_taken: Number of iterations taken to converge
    - history: List of x values during the iterations for visualization
    """
    history = []
    for iteration in range(iterations):
        gradient = 2 * x - 2  # Derivative of f(x) = x^2 - 2x + 2
        x = x - learning_rate * gradient
        history.append(x)

    x_min = x
    iterations_taken = iteration + 1
    return x_min, iterations_taken, history

def plot_function_and_history(history):
    """
    Plot the function and the history of x values during the iterations.

    Parameters:
    - history: List of x values during the iterations
    """
    x_vals = np.linspace(-2, 4, 100)
    y_vals = x_vals**2 - 2*x_vals + 2

    plt.plot(x_vals, y_vals, label='f(x) = x^2 - 2x + 2', color='blue')
    plt.scatter(history, [val**2 - 2*val + 2 for val in history], color='red', label='Gradient Descent Steps')
    plt.title('Gradient Descent for f(x) = x^2 - 2x + 2')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()
```

```python

# Initial guess for the minimum
initial_x = 0

# Hyperparameters
learning_rate = 0.1
iterations = 100

# Perform gradient descent
minima, iterations_taken, history = gradient_descent(initial_x, learning_rate, iterations)

# Display results
print(f"Global Minimum found at x = {minima}")
print(f"Number of Iterations: {iterations_taken}")

# Plot the function and the history of x values during the iterations
plot_function_and_history(history)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent_2d(x, y, learning_rate, iterations):
    """
    Perform gradient descent to find the minimum of the given 2D function.

    Parameters:
    - x: Initial guess for x
    - y: Initial guess for y
    - learning_rate: Step size for each iteration
    - iterations: Number of iterations to perform

    Returns:
    - minima: Tuple (x_min, y_min) representing the values of x and y at the minimum
    - iterations_taken: Number of iterations taken to converge
    - history: List of (x, y) values during the iterations for visualization
    """
    history = []
    for iteration in range(iterations):
        gradient_x = -2 * (1 - x) - 400 * x * (y - x**2)
        gradient_y = 200 * (y - x**2)

        x = x - learning_rate * gradient_x
        y = y - learning_rate * gradient_y

        history.append((x, y))

    minima = (x, y)
    iterations_taken = iteration + 1
    return minima, iterations_taken, history

def plot_2d_function_and_history(history):
    """
    Plot the 2D function and the history of (x, y) values during the iterations.

    Parameters:
    - history: List of (x, y) values during the iterations
    """
    x_vals = np.linspace(-2, 2, 100)
    y_vals = np.linspace(-1, 3, 100)
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = (1 - X)**2 + 100 * (Y - X**2)**2

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

    x_history, y_history = zip(*history)
    z_history = [(1 - x)**2 + 100 * (y - x**2)**2 for x, y in history]
    ax.scatter(x_history, y_history, z_history, color='red', label='Gradient Descent Steps')

    ax.set_title('Gradient Descent for f(x, y) = (1 - x)^2 + 100(y - x^2)^2')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f(x, y)')
    plt.show()
```

```python
# initial guesses for x and y
initial_x = 0.5
initial_y = 0.5

# learning rate
learning_rate = 0.0001

# number of iterations
iterations = 500

# Perform gradient descent
minima, iterations_taken, history = gradient_descent_2d(initial_x, initial_y, learning_rate, iterations)

# Display results
print(f"Global Minimum found at x = {minima[0]}, y = {minima[1]}")
print(f"Number of Iterations: {iterations_taken}")

# Plot the 2D function and the history of (x, y) values during the iterations
plot_2d_function_and_history(history)
```
